```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
titanic=pd.read_csv('train.csv')
titanic.head()
```

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

### 1.Scatterplot (numerical-numerical)

```python
sns.scatterplot(tips['total_bill'],tips['tip'],
                hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

### 2. Barplot(num-cat)

```python
sns.barplot(titanic['Pclass'],titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.barplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

### 3. Boxplot(num-cat)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

### 4. distplot (num - cat)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False) #blue die
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) #orange survived
```

### 5. heatmap

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```

### 6. ClusterMap (Cat-cat)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

### 7.pairplot

```python
sns.pairplot(titanic,hue='Sex')
```

```python
sns.pairplot(iris,hue='species')
```

### 8.Lineplot (num-num)

```python
new=flight.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'],new['passengers'])
```

```python
sns.heatmap(flight.pivot_table(values='passengers',index='month',columns='year'))
```
