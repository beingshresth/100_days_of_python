## Heapmaps

A heat maps is a graphical representation of data where the individual values contained in a matrix are representated as colors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

## ClusterMap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```

```python
sns.clustermap(x,row_cluster=False,annot=True,metric='correlation')
```
