# JointPlot

joinplot creates a multi panel figure that shows both the bivariate(or join)relationship b/w two variable along with the univariate distribution of each on seperate axes.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
```
