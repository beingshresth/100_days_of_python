```python
import pandas as pd
patients=pd.read_csv('patient.csv')
treatment=pd.read_csv('treatment.csv')
adverse=pd.read_csv('adverse_reaction.csv')
```

```python
patients_copy=patients.copy()
treatment_copy=treatment.copy()
adverse_copy=adverse.copy()
```

```python
#3rd problem
treatment_cut=pd.read_csv('treatments_cut.csv')
frames = [treatment , treatment_cut]
treatment_copy=pd.concat(frames)
treatment_copy
```

```python
#2nd problem
treatment_copy=pd.melt(treatment_copy,id_vars=['given_name'
                                               ,'surname'
                                               ,'hba1c_start'
                                               ,'hba1c_end'
                                               ,'hba1c_change']
                       ,var_name='treatment',value_name='dose')
              
```
