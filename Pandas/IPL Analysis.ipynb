```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
# function for most run scored by a batsman against which team
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
run_scored('V Kohli')
```

```python
run_scored('MS Dhoni')
```

```python
run_scored('DA Warner')
```

```python
#isin()
#find the most destructive death over batsman in the history of ipl.
#on the basis of strike rate=(number of runs/number of balls)*100
#min between 200 balls in over 16-20
```

```python
death_over = delivery[delivery['over']>15]
```

```python
#number of balls played by a played in death over
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
all_batsman[x]
```

```python
#run scored by all these 43 batsman
#balls played by these 43 batsman
final=death_over[death_over['batsman'].isin(batsman_list)] 
final_runs=final.groupby('batsman')['batsman_runs'].sum()
final_balls=final.groupby('batsman')['batsman_runs'].count() 
```

```python
sr=(final_runs/final_balls)*100
sr
```
